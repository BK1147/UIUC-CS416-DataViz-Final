```python
# Drop the specified columns vin, condition, interior, seller, mmr, saledate
data = data.drop(columns=['vin', 'condition', 'interior', 'seller', 'mmr', 'saledate'])
data.head(5)
```

```python
data_filtered = data[(data['year'] >= 2010) & (data['odometer'] < 210000)]
data_filtered
```

```python
# Get value counts for the 'body' column
body_counts = data_filtered['body'].value_counts()

# Filter 'body' counts to only include those with more than 20 occurrences
valid_bodies = body_counts[body_counts > 30].index
# valid_bodies
# # Filter the DataFrame to only include rows with valid 'body' values
data_filtered = data_filtered[data_filtered['body'].isin(valid_bodies)]

print(data_filtered)
```

```python
data_filtered['body'] = data_filtered['body'].str.capitalize()
data_filtered['body'].value_counts()
```

```python
data_filtered
```

```python
# Define a function to clean the 'body' column
def clean_body(body):
    if 'coup' in body or 'Koup' in body or 'koup' in body:
        return 'coupe'
    elif 'convertible' in body:
        return 'convertible'
    elif 'cab' in body:
        return 'cab'
    elif 'van' in body:
        return 'van'
    elif 'sedan' in body:
        return 'sedan'
    else:
        return body

data_filtered['body'] = data_filtered['body'].str.lower().apply(clean_body)
data_filtered['body'].value_counts()
```

```python
# Randomly select 4000 rows
df_sampled = data_filtered.sample(n=6000, random_state=1)
df_sampled
```

```python
df_sampled['body'].value_counts()
```

```python
# Define the bins and labels for odometer groups
bins = [0, 20000, 40000, 60000, 80000, 100000, 210000]
labels = ['0-20k', '20k-40k', '40k-60k', '60k-80k', '80k-100k', 'More than 100k']

df_sampled['odometer_custom'] = pd.cut(df_sampled['odometer'], bins=bins, labels=labels, right=False)
df_sampled.head(20)
```

```python
tmp = df_sampled[df_sampled['year'] == 2011]
tmp['odometer_custom'].value_counts()
```
